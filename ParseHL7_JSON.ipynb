```python
import sys
from base64 import b16encode
import codecs
from future.builtins import str, bytes

# Set debug to true to get additional messages when testing or troublshooting
debug = False

# How many messages do you want to loop through? Set to 0 to get all
MessageCount = 0

sAttributeFormat = '{{"{}": "{}"}}'

###############################################################################################

with codecs.open('HL7.txt', 'r',encoding='utf-8', errors='replace') as file_input:
    HL7File = file_input.read()
    
    with open("HL7_File_JSON.txt", "a") as file_output:
        print('{"Messages":[')
        file_output.write(('{"Messages":['))
        ################### Get Messages ############
        iCountMessage = 0
        for Message in HL7File.split("MSH"):
            tmpMessage = Message.encode(encoding='UTF-8',errors='replace')
            if MessageCount !=0:
                if iCountMessage == MessageCount:
                    break;
            PreTxtMsg = "Msg_{}".format(iCountMessage)
            if debug:
                print ("{}_{}\n".format(PreTxtMsg, sys.stdout.bufferMessage.encode(encoding='UTF-8',errors='replace')))
                file_output.write("{}_{}\n".format(PreTxtMsg, sys.stdout.bufferMessage.encode(encoding='UTF-8',errors='replace')))

            iCountMessage += 1


            ################### Get Segments ############        
            iCountSegment = 0
            for Segment in Message.splitlines():
                PreTxtSeg = "{}_Seg{}".format(PreTxtMsg, iCountSegment)
                if debug:
                    print (sAttributeFormat.format(PreTxtSeg, Segment.encode(encoding='ascii',errors='replace')))
                    file_output.write(sAttributeFormat.format(PreTxtSeg, Segment.encode(encoding='ascii',errors='replace')))
                iCountSegment += 1


                ################### Get Fields ############
                iCountField = 0
                for Field in Segment.split("|"):
                    PreTxtFld = "{}_Fld{}".format(PreTxtSeg, iCountField)
                    tmpr = Field.strip()
                    if Field.strip() != "":


                        ################### Get Components ############
                        if (  ("^" in Field) &  (not Field.startswith("^"))  ):
                            #Enters here if there is a carrot symbol and carrot isn't first character
                            iCountComponent = 0
                            for Component in Field.split("^"):
                                PreTxtCom = "{}_Com{}".format(PreTxtFld, iCountComponent)
                                if Component.strip() != "":
                                    print ("    " + sAttributeFormat.format(PreTxtCom, str(Component.encode(encoding='ascii',errors='ignore'))))
                                file_output.write("    " + sAttributeFormat.format(PreTxtCom, str(Component.encode(encoding='ascii',errors='ignore'))))
                                iCountComponent += 1
                        else:
                            #Since no carrot symbols in the field just print the Field
                            if Segment == 0:
                                print("MSH")
                                print ("  " + sAttributeFormat.format(PreTxtFld, str(Field.strip().encode("utf-8"))))
                                file_output.write("  " + sAttributeFormat.format(PreTxtFld, str(Field.strip().encode("utf-8"))))
                            else:
                                print ("  " + sAttributeFormat.format(PreTxtFld, str(Field.strip().encode("utf-8"))))
                                file_output.write("  " + sAttributeFormat.format(PreTxtFld, str(Field.strip().encode("utf-8"))))

                    iCountField += 1

        print(']}')
        file_output.write(']}')

#Sample json
# {"Messages":[
#      {"firstName":"John", "lastName":"Doe"},
#      {"firstName":"Anna", "lastName":"Smith"},
#      {"firstName":"Peter", "lastName":"Jones"}
#  ]}


#MSH - message new line
#Segment - breaks apart , those are separated by new line
#Fields, separated by pipes
#Components separated by carrots (i.e. legal name a pseudonyms)
#Subcomponents sep by tilde (i.e. dosages of drugs, codine)
#Continuation character or repeaters - "\&"

```
